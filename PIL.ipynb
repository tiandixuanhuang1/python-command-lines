```python
from PIL import Image, ImageDraw, ImageFont, ImageFilter

import random

def md_char():
    return chr(random.randint(65, 90))                               #xiao xie字母的ASCll码值在65-90之间

def my_color():                                                              #定义随机的颜色（*， *， *）表示三原色的数值
    return (random.randint(64, 255), random.randint(64, 255), random.randint(64, 255))

def my_color2():
    return (random.randint(32, 127), random.randint(32, 127), random.randint(32, 127))

width = 60*4
height = 60

image = Image.new('RGB', (width, height), (255, 255, 255))

font = ImageFont.truetype("Arial.ttf", 36)

draw = ImageDraw.Draw(image)                                       #构造绘画对象

for i in range(width):
    for j in range(height):
        draw.point((i, j), fill = my_color())                              #转换图像的像素值
        
for t in range(4):
    draw.text((60*t + 10, 10), md_char(), font = font, fill = my_color2())
                                                                                   #draw.text(address, string, font, color) address = (x,y)，表示x坐标与y坐标
    
image = image.filter(ImageFilter.BLUR)                             # 这一步。。。。无语(记住)
image

                                                                                    # import numpy as np
                                                                                    # print(np.array(image)) #使用np中array的方法可以将原来的Image文件转换成numpy中3维数组的形式
```
